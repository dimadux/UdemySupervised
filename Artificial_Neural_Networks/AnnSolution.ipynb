```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import tensorflow
import keras
import sklearn
from keras.models import  Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data.info()
```

```python
y = data["Exited"]
X = data.drop(["Exited","RowNumber","CustomerId","Surname"],axis=1)
```

```python
labelencoder = LabelEncoder()
X["Geography"] = labelencoder.fit_transform(X["Geography"])
X["Gender"] = labelencoder.fit_transform(X["Gender"])
```

```python
X.head()
```

```python
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:,1:]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = Sequential()
classifier.add(Dense(output_dim=20,input_dim=11,activation="relu",kernel_initializer="uniform"))
classifier.add(Dense(output_dim=14,input_dim=20,activation="relu",kernel_initializer="uniform"))
classifier.add(Dense(output_dim=10,input_dim=14,activation="relu",kernel_initializer="uniform"))
classifier.add(Dense(output_dim=10,input_dim=10,activation="relu",kernel_initializer="uniform"))
classifier.add(Dense(output_dim=1,input_dim=10,activation="sigmoid",kernel_initializer="uniform"))
classifier.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
classifier.fit(X_train,y_train, batch_size=10,epochs=100)
```
